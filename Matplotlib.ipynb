```python
import matplotlib as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.plot(years,runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs Scored by Tim")
plt.show()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="virat")
plt.plot(years, sehwag, label="virandra")
plt.title("Runs scored by both batters")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli, label="kohli")
plt.plot(years, sehwag, label="sehwag")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli, 'ro--',label="kohli")
plt.show()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.xlabel("years")
plt.ylabel("runs")
plt.legend()
plt.show()
```
