```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
```

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes_X= diabetes.data
```

```python
print(diabetes_X)
```

```python
diabetes_X_train = diabetes_X[:-30]
diabetes_X_test = diabetes_X[-30:]
```

```python
diabetes_Y_train = diabetes.target[:-30]
diabetes_Y_test = diabetes.target[-30:]
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(diabetes_X_train, diabetes_Y_train)
```

```python
diabetes_Y_predicted = model.predict(diabetes_X_test)
```

```python
print("Mean squared error is: ", mean_squared_error(diabetes_Y_test, diabetes_Y_predicted))
```

```python
print("weights: ",model.coef_)
```

```python
print("Intercept: ",model.intercept_)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(diabetes_Y_test, diabetes_Y_predicted)
plt.plot(
    [diabetes_Y_test.min(), diabetes_Y_test.max()],
    [diabetes_Y_test.min(), diabetes_Y_test.max()]
)

plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values (Linear Regression)")

plt.show()
```

```python
residuals = diabetes_Y_test - diabetes_Y_predicted

plt.figure(figsize=(8, 6))
plt.scatter(diabetes_Y_predicted, residuals)
plt.axhline(y=0)

plt.xlabel("Fitted (Predicted) Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted Values")

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=15)

plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Distribution of Residuals")

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(diabetes_Y_predicted, diabetes_Y_test)

plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.title("Predicted vs Actual Values")

plt.show()
```

```python
X_single = diabetes.data[:, 2]  # BMI feature

plt.figure(figsize=(8, 6))
plt.scatter(X_single[-30:], diabetes_Y_test)

plt.xlabel("BMI Feature")
plt.ylabel("Target Value")
plt.title("Target vs BMI (Test Data)")

plt.show()
```
